<p><b> چرا استاندارد سازی می‌کنیم؟</b></p>
</span>

<span dir="rtl" style="font-family: B Yekan+;">
<p>
استانداردسازی که به عنوان عادی سازی امتیاز z نیز شناخته می شود، یک تکنیک پیش پردازش است که معمولاً در یادگیری ماشین و آمار استفاده می شود. این شامل تبدیل ویژگی های یک مجموعه داده به گونه ای است که آنها میانگین 0 و انحراف استاندارد 1 داشته باشند. در اینجا دلایلی را نشان می دهیم که چرا استانداردسازی مهم است:

1. **مقیاس های یکسان سازی:** ویژگی های مختلف در یک مجموعه داده ممکن است مقیاس ها یا واحدهای اندازه گیری متفاوتی داشته باشند. برخی از الگوریتم‌ها به مقیاس ویژگی‌ها حساس هستند و این می‌تواند منجر به یادگیری جانبدارانه شود. استانداردسازی تضمین می‌کند که همه ویژگی‌ها مقیاس یکسانی دارند و فرآیند یادگیری را منصفانه‌تر می‌کند.

2. **بهبود همگرایی:** بسیاری از الگوریتم‌های بهینه‌سازی مورد استفاده در یادگیری ماشین کارآمدتر عمل می‌کنند و با استانداردسازی ویژگی‌ها سریع‌تر همگرا می‌شوند. این به ویژه برای روش‌های بهینه‌سازی مبتنی بر گرادیان مانند نزول گرادیان صادق است.

3. **افزایش عملکرد مدل:** برخی از الگوریتم‌های یادگیری ماشین، مانند ماشین‌های بردار پشتیبان (SVM) و k-nearest همسایه‌ها (KNN)، مبتنی بر فاصله هستند و می‌توانند تحت تأثیر ویژگی‌های با مقیاس‌های مختلف قرار گیرند. استاندارد کردن ویژگی ها به این الگوریتم ها کمک می کند تا بهتر کار کنند و نتایج دقیق تری تولید کنند.

4. **قابلیت تفسیر:** استانداردسازی می تواند ضرایب یا مقادیر اهمیت ویژگی ها را قابل تفسیرتر کند. هنگامی که ویژگی ها مقیاس های متفاوتی دارند، ضرایب به دست آمده از مدل ها ممکن است مستقیماً تأثیر آنها را بر متغیر هدف منعکس نکنند.

5. **قاعده سازی:** برخی از تکنیک های منظم سازی، مانند منظم سازی L1 و L2، فرض می کنند که همه ویژگی ها در یک مقیاس هستند. استانداردسازی به اعمال منظم سازی به طور موثر کمک می کند و از تسلط برخی ویژگی ها بر فرآیند یادگیری جلوگیری می کند.

6. **مدل هایی که نیاز به توزیع عادی دارند:** برخی از الگوریتم ها، مانند رگرسیون خطی، فرض می کنند که ویژگی ها به طور معمول توزیع شده اند. استانداردسازی می تواند به دستیابی به توزیع نرمال تر در داده ها کمک کند.
</p>
</span>

در این قسمت تمام ویژگی ها را در یک طرف قرار می دهیم و ویژگی هدف را در طرف دیگر قرار می دهیم. ما همچنین داده های آموزشی و داده های تست را به دو قسمت تقسیم می کنیم. سپس ستون های نر و ماده (که در قسمت قبل ایجاد کردیم) را جداگانه بررسی می کنیم.

<h3><b><font color= #BA4A00> Step 29 - Creates a DataFrame </font></b></h3>

The provided code snippet involves the evaluation of various machine learning models on a dataset. Here's what the code does:

1. **Data Splitting:**
   - `X = df_encoded.iloc[:, 1:]`: Extracts the features from the encoded DataFrame, excluding the first column (target).
   - `y = df_encoded.iloc[:, 0]`: Extracts the target variable from the encoded DataFrame.
   - `train_test_split`: Splits the dataset into training and testing sets using a test size of 20% and a random state of 1.
   - Subsequently, the training set is further split into training and validation sets using the same test size and random state.

2. **SMOTE Oversampling:**
   - `smt = SMOTE(random_state=1)`: Creates an instance of the Synthetic Minority Over-sampling Technique (SMOTE) to handle class imbalance.
   - `X_train, y_train = smt.fit_resample(X_train, y_train)`: Applies SMOTE to the training data, generating synthetic samples to balance the class distribution.

3. **Model Evaluation:**
   - A function named `evaluate_model` is defined to evaluate the performance of a given model on training and validation data. Various metrics like accuracy, precision, recall, F1-score, and ROC AUC are calculated for both sets.
   - A dictionary named `models` is defined, containing instances of various machine learning models to be evaluated.

4. **Model Iteration and Evaluation:**
   - A loop iterates through the models in the `models` dictionary.
   - For each model, the `evaluate_model` function is called to obtain the performance metrics for training and validation.
   - The results are stored in a list.

5. **Results Concatenation and Sorting:**
   - After evaluating all models, the individual results are concatenated into a single DataFrame named `results`.
   - The results are sorted in descending order of validation accuracy.

The intention of this code is to assess the performance of different machine learning models on the dataset using various evaluation metrics. The models include Logistic Regression, Decision Tree, Random Forest, K-Nearest Neighbors (KNN), Support Vector Machine (SVM), Naive Bayes, and Gradient Boosting. The results are sorted to identify the model with the highest validation accuracy. This information can guide you in selecting the best-performing model for your specific task.

<font color= #F1C40F> 📌 You can see this code below 👇 </font>

```python
X = df_encoded.iloc[:, 1:]
y = df_encoded.iloc[:, 0]

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=1,
)
X_train, X_val, y_train, y_val = train_test_split(
    X_train,
    y_train,
    test_size=0.2,
    random_state=1,
)
# X_train, X_val, y_train, y_val = train_test_split(
#     X_train, y_train, test_size=0.2, random_state=1
# )
# pd.Series(y_test).value_counts().plot(kind="bar")
smt = SMOTE(random_state=1)
X_train, y_train = smt.fit_resample(X_train, y_train)

def evaluate_model(model, X_train, y_train, X_val, y_val):
    model.fit(X_train, y_train)
    y_train_pred = model.predict(X_train)
    y_val_pred = model.predict(X_val)
    df = pd.DataFrame({'train_accuracy': [accuracy_score(y_train, y_train_pred)],
                        'train_precision': [precision_score(y_train, y_train_pred)],
                        'train_recall': [recall_score(y_train, y_train_pred)],
                        'train_f1': [f1_score(y_train, y_train_pred)],
                        'train_roc_auc': [roc_auc_score(y_train, y_train_pred)],
                        
                        'val_accuracy': [accuracy_score(y_val, y_val_pred)],
                        'val_precision': [precision_score(y_val, y_val_pred)],
                        'val_recall': [recall_score(y_val, y_val_pred)],
                        'val_f1': [f1_score(y_val, y_val_pred)],
                        'val_roc_auc': [roc_auc_score(y_val, y_val_pred)]})
    return df


models = {
    'Logistic Regression': LogisticRegression(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'KNN': KNeighborsClassifier(),
    'SVM': SVC(),
    'Naive Bayes': GaussianNB(),
    'Gradient Boosting': GradientBoostingClassifier()
}

# iterate through the dictionary and evaluate each model
results_l = []
for name, model in models.items():
    model_results = evaluate_model(model, X_train, y_train, X_val, y_val)
    model_results['model'] = name
    results_l.append(model_results)

# concatenate the results
results = pd.concat(results_l, axis=0).reset_index(drop=True)

results.sort_values(by='val_accuracy', ascending=False)
```
